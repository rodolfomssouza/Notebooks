```r
# Gráfico de barras
# Preparando os dados
chuva_normal = dchuva$nc
meses = dchuva$month

par(las=1, family='serif', mar=c(5,5,2,2))
bp1 = barplot(chuva_normal, col='blue', ylim=c(0,200),
              xlab='Meses', ylab='Chuva (mm)')

axis(side=1, at=bp1[,1], labels = meses)
box()
```

```r
# Barras agrupadas
# Preparando os dados
chuvas = t(as.matrix(dchuva[2:4]))
cores = viridis(3)

par(las=1, family='serif', mar=c(5,5,2,2))
bp2 = barplot(chuvas, ylim=c(0,200), xlab='Meses', ylab='Chuva (mm)',
              col=cores, # Cores das barras
              beside=T)  # Agrupar as barras

axis(side=1, at=bp2[2,], labels = meses)
box()

# Legenda
legend('topright', legend=c('Normal', '2012', '2013'),
       fill=cores, bty='n')
```

```r
# Gráfico de barras com outras informações
# Pergunta: Tem diferença entre a temperatura instantânea e a
# média da temperatura máxima e mínima?

# Preparando os dados para análises
tinst = datm$TArmed
tmxmn = (datm$TArmax + datm$TArmin)/2

trats = c(rep('Tinst', length(datm$Dia)), rep('Tmxmn', length(datm$Dia)))
vresp = c(tinst, tmxmn)

# A análise
dic(trat=trats, resp=vresp)
tmps = c(mean(tinst), mean(tmxmn))

par(las=1, family='serif', mar=c(5,5,2,2))
bp3 = barplot(tmps, ylim=c(0,30), ylab='Temperatura (°C)', col='darkred')
axis(side=1, at=bp3[,1], labels=c('Tinst', '(Tmax+Tmin)/2'))

# Adicionar letras do teste de comparação de média
text(x=bp3[,1], y=tmps+1, labels = c('B', 'A'))
box()
```
