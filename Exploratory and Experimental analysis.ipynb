```r
# Data --------------------------------------------------------------------
df1 = read.table('GEOR_2016_120_Data.txt', h=T)
head(df1)
```

```r
# Descriptive statistics --------------------------------------------------
# Summary of the data
summary(df1[3:(length(df1)-1)])
```

```r
# Descriptive statistics (complete)
est = basicStats(df1[3:(length(df1)-1)])
round(est, 4)
```

```r
# Normality test
# Kolmogorov-Smirnov
ks.test(df1$Ks, 'pnorm', mean=mean(df1$Ks), sd=sd(df1$Ks))
```

```r
# Run the test for all variables
pv = NA
interp = NA
for(i in 3:17){
    pv[i-2] = ks.test(df1[,i], 'pnorm',
                        mean=mean(df1[,i]),
                        sd=sd(df1[,i]))$p.value
    if(pv[i-2]>=0.05){interp[i-2]='Normal 5 %'}
    else if(pv[i-2]<0.05 & pv[i-2]>=0.01){interp[i-2]='Normal 1 %'}
    else{interp[i-2]='No normal'}
}
```

```r
# Add p-value of normality test
est[17,] = round(pv, 4)
n1 = rownames(est)
n1[17] = 'PV-KS'
rownames(est) = n1
```

```r
# Save results
write.csv(x = est, file = 'Stats_complete.csv')
```
